```python
#WEEK 2(DATA ENGINEER):DATA CLEANING AND EXPLORATION
data = pd.read_parquet("/content/wildfire_project.xlsx.parquet")
data = data.fillna(data.mean(numeric_only=True))
data.to_parquet("/content/wildfire_project.xlsx.parquet", index=False)
print("Data cleaned and saved!")
```

```python
#CHECKING IF DATA WAS WELL CLEANED(I.E CHECKING DUPLICATES ,DATA TYPES,OUTLIERS )
# Diagnostic Queries to Check Cleaning
print("\n--- Data Cleaning Validation ---")

## 1. Check for Missing Values
missing_values = data.isnull().sum()
print("Missing Values per Column:\n", missing_values)
if missing_values.sum() == 0:
    print("No missing values detected - Cleaning successful!")
else:
    print("Warning: Missing values remain in the dataset.")
```

```python
# 2. Check for Duplicates
duplicates = data.duplicated().sum()
print(f"Number of Duplicate Rows: {duplicates}")
if duplicates == 0:
    print("No duplicates detected - Data is unique.")
else:
    print(f"Duplicates found. Consider removing them with: data = data.drop_duplicates()")
```

```python
# 3. Check Data Types
print("Data Types:\n", data.dtypes)
expected_types = {
    "latitude": "float64", "longitude": "float64", "brightness": "float64",
    "scan": "float64", "track": "float64", "acq_date": "datetime64[ns]",
    "acq_time": "int64", "satellite": "object", "instrument": "object",
    "confidence": "object", "version": "object", "bright_t31": "float64",
    "frp": "float64", "daynight": "object"
}
type_mismatch = {col: dtype for col, dtype in data.dtypes.items() if col in expected_types and str(dtype) != expected_types[col]}
if not type_mismatch:
    print("All data types match expected types.")
else:
    print("Type Mismatch Detected:\n", type_mismatch)
```

```python
# 4. Summary Statistics
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
print("Summary Statistics:\n", data[numeric_cols].describe())
print("Cleaning validation complete!")
```
